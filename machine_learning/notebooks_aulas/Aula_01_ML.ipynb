c) Função que calcula o coeficiente de variação de um arranjo = média/desvio padrão

```python
def CalculaCV(valor):
    cv = valor.std()/valor.mean()
    return cv
```

```python
CV_preco = CalculaCV(preco2)
print(CV_preco)
```

```python
CV_idade = CalculaCV(idade2)
print(CV_idade)
```
